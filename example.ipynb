# LIPSS_SIPE

This notebook is an example to show how the efficacy factor $\eta$ can be calculated.


## Imports

```python
import numpy as np
from lipss_sipe.core import compute_eta_array, plot_eta
```

## Define parameters

$F$ -   Filling factor

$S$ -   Shape factor

$n$ -   Refractive index at the irradiation wavelength

$k$ -   Absorption coefficient at the irradiation wavelength

$\theta$    -   Angle of incidence of the laser in degrees

pol    -   Polarization of the laser, either $s$ or $p$


```python
F = 0.1
S = 10
n = 4  # Ge at 1.06 µm
k = 0.1  # Ge at 1.06 µm
theta = 60
pol = "p"
```

### Wavevector range for $\kappa_x$ and $\kappa_y$

```python
KAPPA_X = np.linspace(-5, 5, 151)
KAPPA_Y = np.linspace(-5, 5, 151)
```

## Calculate $\eta$

```python
if pol == "s":
    POLARIZATION = "spol"
elif pol == "p":
    POLARIZATION = "ppol"
else:
    raise ValueError("pol must be 's' or 'p'")

ETA = compute_eta_array(
    KAPPA_X, KAPPA_Y, POLARIZATION, np.deg2rad(theta), (n + 1j * k) ** 2, F, S
)
```

## Plot $\eta$

```python
plot_eta(ETA, KAPPA_X, KAPPA_Y)
```
